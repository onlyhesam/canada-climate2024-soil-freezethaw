```python
import h5py
import os
import glob
import fnmatch
```

```python
hdf_directory = os.path.join('.', '..', 'data', 'sample')
hdf_directory = os.path.join(
    "/Volumes",
    "Expansion",
    "SateliteBasedData",
    "FT_Product",
    "MEaSUREs_nsidc_0477",
    "Downloads"
)
hdf_files = [os.path.join(hdf_directory, f) for f in os.listdir(hdf_directory) if fnmatch.fnmatch(f, '*CO_FT*_day*.h5')]
```

```python
# Path to your sample HDF file
sample_hdf_path = hdf_files[0]

# Open the HDF file
with h5py.File(sample_hdf_path, 'r') as hdf:
    print("Keys in HDF file:")
    print(list(hdf.keys()))  # Print all keys (datasets) in the HDF file

    # Check the structure and type of 'ft_status' dataset
    if 'ft_status' in hdf:
        ft_status = hdf['ft_status']
        print(list(ft_status))
        print("\nft_status dataset:")
        print(f"Shape: {ft_status.shape}")
        print(f"Data type: {ft_status.dtype}")
        print(f"Sample data: {ft_status[0, 0]}")  # Print a sample value

    # Check the structure and type of 'ft_qc' dataset
    if 'ft_qc' in hdf:
        ft_qc = hdf['ft_qc']
        print(list(ft_qc))
        print("\nft_qc dataset:")
        print(f"Shape: {ft_qc.shape}")
        print(f"Data type: {ft_qc.dtype}")
        print(f"Sample data: {ft_qc[0, 0]}")  # Print a sample value
```
